```python
#b) Remove the units from some of the attributes and only keep the numerical values (for
#example remove kmpl from â€œMileageâ€, CC from â€œEngineâ€, bhp from â€œPowerâ€, and lakh from
#â€œNew_priceâ€). (5 points)
```

```python
cars_data_copy = cars_data.copy()
cars_data_copy['Mileage'] = cars_data_copy['Mileage'].astype(str).str.replace('kmpl', '').str.replace('km/kg', '').astype(float)
cars_data_copy['Engine'] = cars_data_copy['Engine'].astype(str).str.replace('CC', '').astype(float)
cars_data_copy['Power'] = cars_data_copy['Power'].astype(str).str.replace('bhp', '').astype(float)
cars_data_copy['New_Price'] = cars_data_copy['New_Price'].astype(str).str.replace('Lakh', '')
cars_data_copy['New_Price'] = cars_data_copy['New_Price'].str.replace('Cr', '').apply(lambda x: float(x) * 100 if 'Cr' in x else float(x))
```

```python
cars_data_copy
```
